```python
import numpy as np        # Numerical python uesd for ndarray 
import pandas as pd       # I/O , data manipulation 
import matplotlib.pyplot as plt     #numpy, pandas, matplotlib are used for data processing
from pandas.plotting import scatter_matrix
from sklearn.model_selection import train_test_split     # Divide data into Training and test dataset
from sklearn import metrics      # Used for accuracy of our model
from sklearn.neighbors import KNeighborsClassifier 
from sklearn.linear_model import LogisticRegression
```

```python
# Load Dataset & see top five rows
data = pd.read_csv('iris.csv')
data.head()
```

```python
# Remove unneccessary data
data.drop('Unnamed: 0', axis = 1, inplace = True)
```

```python
# Check for null values in our dataset
data.isnull().sum()
```

```python
# Quick statictics overview of data
data.describe()
```

```python
# Check Dependent variable how many unique values in it
data['Species'].unique()
```

```python
# convert character in numeric form
data['Species'] = data['Species'].map({'setosa':0, 'versicolor':1, 'virginica':2})

# Select independent and dependent variable
X = data.iloc[:, :-1]
Y = data.iloc[:, -1]

# Split data into train(80%) & test(20%) data 
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size = 0.2, shuffle = True)

# Here our model is used for train the model through training data and make prediction 
knn = KNeighborsClassifier(n_neighbors = 5)
knn.fit(x_train, y_train)
y_pred = knn.predict(x_test)

# Check accuracy score
print("Accuracy",metrics.accuracy_score(y_test, y_pred))
```
